```python
import yfinance as yf

# Get the current date
current_date = datetime.date.today()

# Calculate the year-to-date gain for META and TESLA
meta = yf.Ticker('META')
tesla = yf.Ticker('TSLA')

tesla.get_history_metadata()

ydgain_meta = (current_date - meta.historical_prices['close'][0]) / meta.historical_prices['close'][0] * 100
ydgain_tesla = (current_date - tesla.historical_prices['close'][0]) / tesla.historical_prices['close'][0] * 100

print('Year-to-date gain for META:', ydgain_meta)
print('Year-to-date gain for TESLA:', ydgain_tesla)
```
