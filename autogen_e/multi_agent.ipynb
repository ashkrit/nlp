```python
import requests
from bs4 import BeautifulSoup

# Set the Wikipedia URL for Amazon
url = "https://en.wikipedia.org/wiki/Amazon"

# Send a GET request to the Wikipedia URL and get the HTML response
response = requests.get(url)

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(response.content, "html.parser")

print(soup.prettify())

# Find the search bar on the Wikipedia page
search_bar = soup.find("input", name="search",placeholder="Search Wikipedia")

# Perform a search for "Amazon" on Wikipedia
search_results = search_bar.value

# Print the search results
print(search_results)
```
